```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Parabola")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, "b")

plt.subplot(1,2,2)
plt.plot(y,x,"r")


plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Parabola")
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes2.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.15, 0.4, 0.3])
axes3 = fig.add_axes([0.3, 0.5, 0.4, 0.3])

axes1.set_title("Main Axes")
axes2.set_title("Axes 2")
axes3.set_title("Axes 3")

axes1.plot(x,y)
axes2.plot(y,x)
axes3.plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2)
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(x,y,label="X Squared")
axes[1].plot(y,x,label="Root")

fig.legend()

plt.tight_layout()
```
